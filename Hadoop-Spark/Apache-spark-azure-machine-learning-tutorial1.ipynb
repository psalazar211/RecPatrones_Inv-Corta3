## Crear un grupo de Apache Spark sin servidor usando Synapse Studio
Azure Synapse Analytics ofrece varios motores de análisis para ayudarle a ingerir, transformar, modelar, analizar y servir sus datos. El grupo Apache Spark ofrece capacidades informáticas de big data de código abierto. Después de crear un grupo de Apache Spark en su espacio de trabajo de Synapse, los datos se pueden cargar, modelar, procesar y servir para obtener información. Se utilizará el espacio de trabajo de Synapse.

## Create the Apache Spark pool in Synapse Studio

1. En la página de inicio de Synapse Studio, navegue hasta Management Hub en la navegación izquierda seleccionando el ícono Administrar.

2. Una vez en Management Hub, navegue hasta la sección Grupos de Apache Spark para ver la lista actual de grupos de Apache Spark que están disponibles en el espacio de trabajo.

```python
from IPython.display import Image
Image("Images/new_apache_spark_pool.png")
```

3. Seleccione + Nuevo y aparecerá el nuevo asistente de creación del grupo Apache Spark.

4. Ingrese los siguientes detalles en la pestaña Conceptos báicos:

| Configuración | Valor sugerido | Descripción |
|-----------------------|-----------------------|-----------------------|
|Nombre del grupo de Apache Spark | Un nombre de grupo válido, como contosospark. | Este es el nombre qu tendrá el grupo de Apache Spark |
|Tamaño de nodo |Pequeño (4 vCPU / 32 GB) | Configúrelo en el tamaño más pequeño para reducir los costos de este inicio rápido |
|Escala automática | Deshabilitada | No necesitaremos escala automática en este inicio rápido |
|Número de nodos |4 |Utilice un tamaño pequeño para limitar los costos en este inicio rápido |
|Asignar ejecutores dinámicamente | Deshabilitado | Esta configuración se asigna a la propiedad de asignación dinámica en la configuración de Spark para la asignación de ejecutores de aplicaciones Spark. No necesitaremos el escalado automático en este inicio rápido. |

```python
from IPython.display import Image
Image("Images/new_apache_spark_pool_config.png")
```

5. En la siguiente pestaña, Configuración adicional, deje todas las configuraciones como predeterminadas.

6. Seleccione Etiquetas. Considere la posibilidad de utilizar etiquetas de Azure. Por ejemplo, la etiqueta "Propietario" o "Creado por" para identificar quién creó el recurso y la etiqueta "Entorno" para identificar si este recurso está en producción, desarrollo, etc. Para obtener más información, consulte Desarrollar su estrategia de nomenclatura y etiquetado para Recursos de Azure. Cuando esté listo, seleccione Revisar + crear.

7. En la pestaña Revisar + crear, asegúrese de que los detalles se vean correctos según lo que se ingresó anteriormente y presione Crear.

8. El grupo de Apache Spark iniciará el proceso de aprovisionamiento.

9. Una vez que se complete el aprovisionamiento, el nuevo grupo de Apache Spark aparecerá en la lista.
